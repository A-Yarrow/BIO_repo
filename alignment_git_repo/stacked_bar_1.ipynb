```python
import pandas as pd
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline



filename = 'heavy_chain_nr_CH2_CH3_aa_pct'
df = pd.read_csv(filename, sep='\t', header=2)
df.set_index('Position', inplace=True)
del df['GAP']
position_updated = [i+225 for i in range(238)]
df['position_updated'] = position_updated

my_colors = ['#b3b3ff', '#99ebff', '#c2d6d6', '#9fdf9f', '#ffffb3', '#d6d6c2', '#ffd480', '#ffbf80', '#80ffff', '#4d4dff', '#ff99ff', '#bfff80', '#cccccc', '#99ffcc', '#b3ccff', '#6666ff', '#dfbf9f', '#dfff80']

interaction_aa = [252, 253, 254, 309, 310, 311, 312, 314, 315, 433, 435, 436]
df_interaction = df[df['position_updated'].isin(interaction_aa)]
df_new_index = df_interaction.set_index('position_updated')
df_new_index.head()

fig = plt.figure()
ax = df_new_index.plot(kind='bar', stacked = True, color=my_colors)
plt.legend(loc ='right', bbox_to_anchor=(1.3, 0.625))
fig.set_size_inches(5, 3)
fig1 = plt.gcf()
fig.show()
fig1.savefig('CH2_CH3_stacked_bar.png', dpi=300)
```
